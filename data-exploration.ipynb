```python
import pandas as pd
import geopandas as gpd

# Define file paths (update these paths based on your local directory)
file_paths = {
    "listings": "data/listings.csv.gz",
    "calendar": "data/calendar.csv.gz",
    "reviews": "data/reviews.csv.gz",
    "listings_summary": "data/listings.csv",
    "reviews_summary": "data/reviews.csv",
    "neighbourhoods": "data/neighbourhoods.csv",
    "neighbourhoods_geojson": "data/neighbourhoods.geojson"
}

# Load files into individual dataframes
listings = pd.read_csv(file_paths["listings"], low_memory=False)
calendar = pd.read_csv(file_paths["calendar"], low_memory=False)
reviews = pd.read_csv(file_paths["reviews"], low_memory=False)
listings_summary = pd.read_csv(file_paths["listings_summary"], low_memory=False)
reviews_summary = pd.read_csv(file_paths["reviews_summary"], low_memory=False)
neighbourhoods = pd.read_csv(file_paths["neighbourhoods"], low_memory=False)
neighbourhoods_geojson = gpd.read_file(file_paths["neighbourhoods_geojson"])

# Display summary and head of each dataframe
dataframes = {
    "listings": listings,
    "calendar": calendar,
    "reviews": reviews,
    "listings_summary": listings_summary,
    "reviews_summary": reviews_summary,
    "neighbourhoods": neighbourhoods,
    "neighbourhoods_geojson": neighbourhoods_geojson
}

for df_name, df in dataframes.items():
    print(f"DataFrame name: {df_name}")
    print(f"Summary of {df_name}:")
    print(df.info())
    print(f"\nFirst 5 rows of {df_name}:")
    print(df.head())
    print("\n")
```

Data Exploration


```python
# Drop columns with all NaN values
for df_name, df in dataframes.items():
    df.dropna(axis=1, how='all', inplace=True)

# Fill NaN values with appropriate strategies
dataframes['listings'].fillna({
    'price': dataframes['listings']['price'].median(),
    'reviews_per_month': 0,
    'last_review': 'No reviews',
    'license': 'No license'
}, inplace=True)

dataframes['calendar'].fillna({
    'price': dataframes['calendar']['price'].mode()[0],
    'adjusted_price': dataframes['calendar']['adjusted_price'].mode()[0]
}, inplace=True)

dataframes['reviews'].fillna('No comments', inplace=True)
```

```python
# Summary statistics
print(dataframes["listings"].describe())
print(dataframes["calendar"].describe())
print(dataframes["reviews"].describe())

# Check for missing values
print(dataframes["listings"].isnull().sum())
print(dataframes["calendar"].isnull().sum())
print(dataframes["reviews"].isnull().sum())
```

Step 4: Clustering Listings
Prepare Data for Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Select relevant columns for clustering
features = ["latitude", "longitude", "price", "minimum_nights", "number_of_reviews", "reviews_per_month"]
listings_clustering = dataframes["listings"][features].dropna()

# Standardize the features
scaler = StandardScaler()
listings_scaled = scaler.fit_transform(listings_clustering)
```

Apply KMeans Clustering

```python
# Determine the optimal number of clusters using the elbow method
import matplotlib.pyplot as plt

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(listings_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Fit KMeans with the optimal number of clusters (e.g., 4)
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=42)
listings_clustering['cluster'] = kmeans.fit_predict(listings_scaled)

# Add cluster labels to the original dataframe
dataframes["listings"] = dataframes["listings"].merge(listings_clustering[['cluster']], left_index=True, right_index=True)
```

Step 5: Segmenting Hosts
Prepare Data for Segmentation

```python
# Group by host_id and aggregate relevant features
hosts = dataframes["listings"].groupby("host_id").agg({
    "id": "count",
    "number_of_reviews": "sum",
    "reviews_per_month": "mean",
    "calculated_host_listings_count": "mean"
}).rename(columns={"id": "listing_count"})

# Standardize the features
hosts_scaled = scaler.fit_transform(hosts)
```

Apply KMeans Segmentation

```python
# Determine the optimal number of clusters using the elbow method for hosts
wcss_hosts = []
for i in range(1, 11):
    kmeans_hosts = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans_hosts.fit(hosts_scaled)
    wcss_hosts.append(kmeans_hosts.inertia_)

plt.plot(range(1, 11), wcss_hosts)
plt.title('Elbow Method for Hosts')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# Fit KMeans with the optimal number of clusters (e.g., 3)
kmeans_hosts = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
hosts['segment'] = kmeans_hosts.fit_predict(hosts_scaled)

# Add segment labels to the original listings dataframe
dataframes["listings"] = dataframes["listings"].merge(hosts[['segment']], left_on="host_id", right_index=True)
```

Step 6: List Top Three Factors Impacting the Listing Price


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix
correlation_matrix = dataframes["listings"][["price", "latitude", "longitude", "minimum_nights", "number_of_reviews", "reviews_per_month", "calculated_host_listings_count"]].corr()

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Factors impacting the listing price
top_factors = correlation_matrix['price'].sort_values(ascending=False)[1:4]
print("Top three factors impacting the listing price:")
print(top_factors)

# Descriptions:
# 1. `calculated_host_listings_count`: Higher number of listings by the same host can indicate better management and reputation, potentially leading to higher prices.
# 2. `reviews_per_month`: Listings with more frequent reviews can indicate higher demand and popularity, which can justify a higher price.
# 3. `number_of_reviews`: A higher number of reviews can signal trust and reliability to potential guests, often resulting in higher pricing power.
```
